```python
#  "ExpenseTracker Pro"

import re
import getpass
import sys

# new userid check

def USERCHECK(USER):
            with open("USERS_DETAILS.txt","r") as a:
                for line in a:
                    stored_name,_ = line.strip().split(",")
                    if USER==stored_name:
                        return False
            return True

# repeat process
def repeat_process():
    while True:
                user_choice = After_log_in()
                if user_choice == "1":
                    Expenses_Entry()
                elif user_choice == "2":
                    Delete_Expenses()
                elif user_choice == "3":
                    Expenses_Summery()
                elif user_choice == "4":
                    sys.exit()
                else:
                    print("Invalid choice. Please enter a valid option.")

#New User Creating Account

def New_User():
    while True:
    
        USER=input("enter user id: ")
        if USERCHECK(USER):
            print("user id available")
            break
        else:
            print("user id alredy taken")
        
    
    
    PASSWORDPATTERN=r"^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[!@#$%^&*()-_+=])[A-Za-z\d!@#$%^&*()-_+=]+$"  
    
    print("\nAt least one uppercase letter, At least one lower letter, At least one digit, At least one special character\n")

    while True:
        PASSWORD=getpass.getpass("Enter password: ")
        if re.fullmatch(PASSWORDPATTERN,PASSWORD):
            REPASSWORD=getpass.getpass("\nRe-Enter the password: ")
            if PASSWORD == REPASSWORD:
                print("\nPassword matched\n")

                print("USER CREATED SUCCESSFULLY. Thank you\n")
                try:
                    with open ("USERS_DETAILS.txt","a") as a:
                        a.write(f"{USER},{PASSWORD}\n")

                    with open(f"{USER}.txt","w"):
                        pass

                except (IOError, PermissionError) as e:
                    print(f"\nError writing to file: {e}\n")
                break
            else:
                print("\nPassword Not-Matching\n")

        else:
            print("\nfollow password patten\n")

        
# Existing User Login Function

def LOGIN(USERID,PASSWORD):
        with open("USERS_DETAILS.txt","r") as a:
            for line in a:
                check_user_id,check_password=line.strip().split(",")
                if USERID == check_user_id and PASSWORD == check_password:
                    return True
        return False
    

#after login entryes

def After_log_in():
    print("\n1-Entry Expenses")
    print("2-Delete Expenses")
    print("3-Summary")
    print("4-Log_Out")

    return input("\nenter your option: ")



#for Expenses Entry

def Expenses_Entry():
    Date_formate=re.compile(r"^\d{2}-\d{2}-\d{4}")
    Date=input("\nenter data DD/MM/YYYY\n")
    
    if Date_formate.match(Date):
        
        Category=input("enter Category\n").lower()
        Amount=input("enter the amount\n")
        
        try:
            with open (f"{USERID}.txt","a") as a:
                a.write(f"{Date},{Category},{Amount}\n")
            
                print("Expense entry recorded successfully.")
            
        except (IOError, PermissionError) as e:
            print(f"Error writing to file: {e}")
            
        
    else:
        print("Invalid Date")
    
    while True:
        user_choice = input("\nDo you want to:\n1. Enter another expense\n2. Go back to main menu\n\nEnter your choice: ")
        if user_choice == "1":
            Expenses_Entry()
        elif user_choice == "2":
            repeat_process()
        else:
            print("\nInvalid choice. Please enter 1 or 2.\n")

       

#for delete Expenses

def Delete_Expenses():
    Category_for_delete=input("\nCatogory you want to delete: ").lower()
    
    try:
        with open(f"{USERID}.txt","r") as a:
            All_Expenses_in_list=a.readlines()

        Updated_Expenses=[]
        for Category in All_Expenses_in_list:
            if Category_for_delete not in Category.lower():
                Updated_Expenses.append(Category)
                
        if len(All_Expenses_in_list) == len(Updated_Expenses):
            print(f"\nNo expenses found for category {Category_for_delete}. Nothing deleted\n")
        
        else:
            print("\nExpenses Updated Successfully.\n")

        with open(f"{USERID}.txt","w") as a:
            a.writelines(Updated_Expenses)
    
    except FileNotFoundError:
        print(f"\nNo expenses data found for user {USERID}.\n")
    except Exception as e:
        print(f"\nAn error occurred: {e}\n")
        
#for Expenses Summery

def Expenses_Summery():
    Total_Expenses=0
    try:
        with open(f"{USERID}.txt","r")  as a:
            aa=a.readlines()
            
        if not aa:
            print(f"\nNo expense data found for user {USERID}.\n")
            return
        for lists in aa:
            _,_,amount=lists.strip().split(",")
            Total_Expenses+=eval(amount)
        
        print(f"\nYour Total Expenses Rs: {Total_Expenses}/-\n\n")
        
        Expenses_Comp=input("Do you want to Calculate with your Income (y/n)\n")
    
        if Expenses_Comp=="y":
            Expenses_Comparision(Total_Expenses)
        elif Expenses_Comp=="n":
            print("\nThank Your. Your Expenses are updated\n")
        else:
            print("\ninvalid")
        
    except FileNotFoundError:
        print(f"\nNo expense data found for user {USERID}.")
    except Exception as e:
        print(f"\nAn error occurred: {e}")
        

        
#comparing expenses with Income

def Expenses_Comparision(Total_Expenses):
    try:
        
        Income=float(input("\nenter your income: "))
    except ValueError:
        print("\nInvalid input. Please enter a valid number.")
        return
    
    if Income > Total_Expenses:
        saving=Income-Total_Expenses
        print(f"\nYou saved Rs: {saving}/- Good ")
    elif Income < Total_Expenses:
        saving=abs(Income-Total_Expenses)
        print(f"\nYour Expenses is  Rs: {saving}/- More than your Income ")
        
    elif Income == Total_Expenses:
        print("\nYou spend all amount for Expenses Nothing Left for you")
        
    else:
        print("\nSomething is wrong")
    
def exit():
    return print("\nThank You. Visit Again.")


while True:    
    print("\nMENU:\n1. New User\n2. Existing User\n3. Exit-3\n")
    INPUT=input()
    if INPUT=="1":
        New_User()
    elif INPUT=="2":

        max_login_attempts = 3
        login_attempts = 0

        while login_attempts < max_login_attempts:
            USERID=input("Enter User id: ")
            PASSWORD=getpass.getpass("Enter Password: ")

            if LOGIN(USERID, PASSWORD):
                print("Login successful. Welcome!\n")

                while True:
                    user_choice = After_log_in()
                    if user_choice == "1":
                        Expenses_Entry()
                    elif user_choice == "2":
                        Delete_Expenses()
                    elif user_choice == "3":
                        Expenses_Summery()
                    elif user_choice == "4":
                        print("Log out succefully")
                        break
                    else:
                        print("Invalid choice. Please enter a valid option.")
                break
            else:
                print("UserID or Password is not correct. Please try again.")
                login_attempts += 1

                if login_attempts == max_login_attempts:
                    print("Maximum login attempts reached. Exiting program.")
                    break
                    
    elif INPUT=="3":
        print("Session Completed")
        break
        



```
